```python
#Q3
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
paragraph = [
    "The vibrant hues and bold strokes in this painting evoke a powerful sense of emotion. Its abstract nature invites personal interpretation, making it a truly engaging piece of art.",
    "Another Friday night alone. My attempts at finding someone special have been consistently unsuccessful. Perhaps my standards are too high, or maybe I'm just incredibly unlucky in love.",
    "My friends are the best support system I could ask for. Their laughter is infectious, and their unwavering encouragement helps me through tough times. I'm so grateful for their friendship."
]


count_vectorizer = CountVectorizer(stop_words='english')
bow = count_vectorizer.fit_transform(paragraph)
bow_df = pd.DataFrame(bow.toarray(), columns=count_vectorizer.get_feature_names_out())
print(bow_df)
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(paragraph)
tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print(tfidf_df)
```

```python
for i, row in enumerate(tfidf_df.values):
    top_indices = row.argsort()[-3:][::-1]
    top_terms = [(tfidf_df.columns[idx], row[idx]) for idx in top_indices]
    print(f"\nText {i+1}:")
    for word, score in top_terms:
        print(f"  {word}: {score:.4f}")
```

```python
#Q4


import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
text1 = """Manga is basically Japanese comic books or graphic novels. They're usually in black and white and are read from right to left (yep, the opposite of how we read in the West). Manga covers everything—action, romance, fantasy, horror, sports, you name it. People of all ages in Japan read manga, and it's a huge part of their culture. Some of your favorite anime probably started as a manga first!"""
text2 = """Anime is Japanese animation—like cartoons, but with way more variety in tone, style, and story. From epic battles to emotional dramas to hilarious comedies, anime has it all. It’s super expressive, colorful, and often adapts manga into motion. Whether it's short series or long-running shows, anime has become a worldwide obsession for good reason!"""
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text.split()

tokens1 = preprocess(text1)
tokens2 = preprocess(text2)
set1 = set(tokens1)
set2 = set(tokens2)
jaccard_sim = len(set1.intersection(set2)) / len(set1.union(set2))

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform([text1, text2])
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

print(f"Jaccard Similarity: {jaccard_sim:.4f}")
print(f"Cosine Similarity: {cosine_sim:.4f}")
```

```python
#Q5

from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

review = "I had a fantastic stay at this hotel! The room was spotless, modern, and incredibly comfortable. The bed felt like sleeping on a cloud, and I loved the little touches—plenty of charging ports, a smart TV, and great lighting. The bathroom was clean and well-stocked with high-quality toiletries. Staff were super friendly and always ready to help. Plus, the location was perfect—close to everything I needed. I’d absolutely stay here again!"

blob = TextBlob(review)
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity


print(blob)
```

```python
#Q6

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np


paragraph = "A trade war happens when countries start placing tariffs or other restrictions on each other's goods and services in response to trade disputes. It's like an economic tug-of-war, where each side tries to protect its own industries by making foreign products more expensive. While the goal is often to boost local businesses, trade wars can backfire—raising prices for consumers, disrupting global supply chains, and straining international relations. One well-known example is the U.S.-China trade war, which affected everything from electronics to agriculture. In the long run, trade wars usually create more uncertainty than stability in the global economy."
tokenizer = Tokenizer()
tokenizer.fit_on_texts([paragraph])
total_words = len(tokenizer.word_index) + 1


input_sequences = []
for line in paragraph.split('.'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

max_sequence_len = max([len(seq) for seq in input_sequences])
padded_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')


x, y = padded_sequences[:,:-1], padded_sequences[:,-1]
y = np.eye(total_words)[y]


model = Sequential([
    Embedding(total_words, 100, input_length=max_sequence_len-1),
    LSTM(150),
    Dense(total_words, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x, y, epochs=100, verbose=0)
```

```python
def generate_text(seed_word, num_words):
    generated_text = seed_word
    for _ in range(num_words):
        token_list = tokenizer.texts_to_sequences([seed_word])[0]
        padded_token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted_probabilities = model.predict(padded_token_list, verbose=0)[0]
        predicted_index = np.argmax(predicted_probabilities)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_index:
                output_word = word
                break
        generated_text += " " + output_word
        seed_word = output_word
    return generated_text


seed_word = "rain"
num_lines_to_generate = 3


for _ in range(num_lines_to_generate):
    generated_line = generate_text(seed_word, 8)


print(generated_line)
seed_word = generated_line.split()[-1]
```
