```python
# Q1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(102317193)
```

```python
# Part I
sales_data = np.random.randint(1000, 5001, size=(12, 4))
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df = pd.DataFrame(sales_data, columns=columns, index=months)
print(df)
```

```python
# Part II
print("First 5 rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe())
```

```python
totalSalesCategory = df.sum()
print("\nTotal Sales per Category:\n", totalSalesCategory)
```

```python
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

roll_number = 102317156
if roll_number % 2 == 0:
    df["Electronics"] *= 0.9
else:
    df["Clothing"] *= 0.85

print(df)
```

```python
# Part III:
plt.figure(figsize=(10, 6))
sns.lineplot(data=df[columns])
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend(columns)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df[columns])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])
flatArray = array.flatten()
absoluteValues = np.abs(array)
percentilesFlattened = np.percentile(flatArray, [25, 50, 75])
percentilesColumns = np.percentile(array, [25, 50, 75], axis=0)
percentilesRows = np.percentile(array, [25, 50, 75], axis=1)

meanFlattened = np.mean(flatArray)
meanColumns = np.mean(array, axis=0)
meanRows = np.mean(array, axis=1)

medianFlattened = np.median(flatArray)
medianColumns = np.median(array, axis=0)
medianRows = np.median(array, axis=1)

stdFlattened = np.std(flatArray)
stdColumns = np.std(array, axis=0)
stdRows = np.std(array, axis=1)

print("Absolute Values:\n", absoluteValues)
print("Percentiles (Flattened):\n", percentilesFlattened)
print("Percentiles (Columns):\n", percentilesColumns)
print("Percentiles (Rows):\n", percentilesRows)

print("Mean (Flattened):\n", meanFlattened)
print("Mean (Columns):\n", meanColumns)
print("Mean (Rows):\n", meanRows)

print("Median (Flattened):\n", medianFlattened)
print("Median (Columns):\n", medianColumns)
print("Median (Rows):\n", medianRows)

print("Standard Deviation (Flattened):\n", stdFlattened)
print("Standard Deviation (Columns):\n", stdColumns)
print("Standard Deviation (Rows):\n", stdRows)
```
